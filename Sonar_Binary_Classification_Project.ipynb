# Sonar_Binary_Classification_Project
- importing all of the classes and functions we will need:

```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

- Next, we can initialize the random number generator to ensure that we always get the same results when executing this code. This will help if we are debugging:

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

- Load the dataset using pandas 
- Split the columns into 60 input variables (X) and 1 output variable (Y)

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
```

```python
# load dataset
dataframe = pandas.read_csv(io.BytesIO(uploaded['sonar.csv']), header=None)
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:60].astype(float)
Y = dataset[:,60]
```

- Values in variables

```python
X
```

```python
Y
```

- The output variable is string values. We must convert them into integer values 0 and 1.

- We can do this using the LabelEncoder class from scikit-learn.

```python
le = LabelEncoder()
```

```python
le.fit(Y)
encoded_Y = le.transform(Y)
```

```python
encoded_Y
```

- To use Keras models with scikit-learn, we must use the KerasClassifier wrapper. This class takes a function that creates and returns our neural network model. It also takes arguments that it will pass along to the call to fit() such as the number of epochs and the batch size.


```python
# baseline model
def create_baseline():
    # create model
    model = Sequential()
    model.add(Dense(60, activation='relu', input_shape=(60,)))
    model.add(Dense(1, activation = 'sigmoid'))
    # Compile model
    model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'])
    return model
```

- Now it is time to evaluate this model using stratified cross validation in the scikit-learn framework

```python
# evaluate model with standardized dataset
estimator = KerasClassifier(build_fn=create_baseline, epochs=100, batch_size=5, verbose=0)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(estimator, X, encoded_Y, cv=kfold)
print("Results: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

<h3> Re-Run The Baseline Model With Data Preparation

```python
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn=create_baseline, epochs=100, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, encoded_Y, cv=kfold)
print("Standardized: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

Running this example provides the results below. We do see a small but very nice lift in the mean accuracy.

#  Rewriting the code by doing Model Subclassing

```python
import keras
from keras import layers
class Subclassing_Model (keras.Model):
  def __init__(self):
    super (Subclassing_Model, self). __init__()
    self.dense1 = layers.Dense(60, activation='relu')
    self.dense2 = layers.Dense(30, activation='relu')
    self.dense3 = layers.Dense(30, activation='relu')
    self.dense4 = layers.Dense(1, activation='sigmoid')
    
  def call(self,inputs):
    x = self.dense1(inputs)
    x = self.dense2(x)
    x = self.dense3(x)
    return self.dense4(x)

def subclassing():
# create model
    model = Subclassing_Model()
    # Compile model
    model.compile(optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['accuracy'])
    return model
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn=subclassing, epochs=200, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, encoded_Y, cv=kfold)
print("Larger: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
